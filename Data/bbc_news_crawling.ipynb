```python
from bs4 import BeautifulSoup as bs
#from selenium import webdriver
#from selenium.webdriver.common.by import By
# from selenium.webdriver.common.keys import keys
import requests
import urllib.request

import time 


 # 출력 파일 명
OUTPUT_FILE_NAME = 'bbc news.txt'

def get_text(URL):
    print(URL+'='*20)
    source_code_from_URL = urllib.request.urlopen(URL)
    soup = BeautifulSoup(source_code_from_URL, 'lxml', from_encoding='utf-8')
    text = ''
#    print(len(soup.find_all('div', {'class':'GridItemConstrainedMedium-sc-12lwanc-2 fVauYi'})))
    for item in soup.find_all('div', {'class':'GridItemConstrainedMedium-sc-12lwanc-2 fVauYi'}):
        text = text + str(item.text)
    
    #print(text)
    return text

def main(filename, url):
    open_output_file = open(filename, 'a')
    result_text = get_text(url)
    open_output_file.write(result_text)
    open_output_file.close()
    

# driver = wd.Chrome(executable_path='/Users/iris/Downloads/chromedriver')
# driver.get(url)


 # 출력 파일 명
#OUTPUT_FILE_NAME = 'bbc news.txt'
# 긁어 올 URL
#URL = 'https://www.bbc.com/thai/thailand-53469936'

for page in range(1,97):
    # 뉴스 사이트 url 다시 만들기
    url = 'https://www.bbc.com/thai/topics/cjgn73g98rqt/page/'+str(page)
    web = requests.get(url).content
    source = bs(web,'html.parser')


    urls_list = []
    for urls in source.find_all('a', {'class':'qa-heading-link'}):
        new_url_content = urls["href"]
        new_url = "https://www.bbc.com"+new_url_content

        urls_list.append(new_url)
    
    urls_list

 

    for url in urls_list:
        main(OUTPUT_FILE_NAME, url)    

    print(str(page)+"=============")
print("완료")
# https://www.bbc.com/thai/topics/cjgn73g98rqt/page/2


# contents_list = []

# for i in range(96):
#     driver.get(urls_list[i])
#     print(driver.current_url)
    
#     time.sleep(3)
#     try:
#         # 페이지 클릭
#         while driver.find_element_by_xpath('//*[@id="lx-stream"]/div/div/div[3]/a[1]').text !='':
#             driver.find_element_by_xpath
            
```
