```python
pattern1 = re.compile('[{}■「※」ㆍ√☎㎡□○◦㎥･◆￦￭▸▣♣◇←－：…………☞≫≪Ⅰ《》〈〉∨…➀Ⅲ➁‣⇨➂∙▪ⓛ‣Ⅱ․․․▲△Ｏ•×【】▶✓➡◀∼″→ㅇ⇒『』◯▼㎜×·④①②③⑤⑦⑥⑧⑨⑩⑪⑫⑬⑭⑮⑯⑰⑱｢｣◎❶❷❸❹❼❺❻]'.format(re.escape(punctuation)))
pattern2 = re.compile(r'\bwww\w+(?:[.]?\w+)+\b')
pattern3 = re.compile(r'[A-Za-z一-龥0-9]')
pattern4 = re.compile(r'\s{2,}')
```

```python
with open('total_text.txt', 'r',encoding='utf-8') as file:
    for text in file:
        print(text.strip('\n'))
        
```

```python
with open('total_text.txt', 'r',encoding='utf-8') as file:
    for text in file:
        doc = pattern1.sub(' ',
              re.sub(r' {2,}','', 
              pattern2.sub('',
              pattern3.sub('',
              pattern4.sub('',text.strip())))))
        print(doc)
```

```python
import zipfile
```

```python
def read_input(zip_data_file_name, data_file_name):
    with zipfile.ZipFile(zip_data_file_name).open(data_file_name) as f:
        for i, line in enumerate(f):
            yield line.strip()
            
zip_data_file_name = 'total_text.zip'
data_file_name = 'total_text.txt'
documents = list(read_input(zip_data_file_name, data_file_name))
```

```python
import time
for line in read_input(zip_data_file_name, data_file_name):
    time.sleep(2)
    print(line)
    break
```
