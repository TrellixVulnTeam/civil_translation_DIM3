```python
from sqlalchemy.orm import sessionmaker
```

```python
engine = create_engine('sqlite:///min_corpus.db', echo=True)
```

```python
base = declarative_base()
```

```python
class MinwonCorpus(base):
    __tablename__ = 'MINWON_CORPUS'
    
    pk = Column('pk', Integer, primary_key=True, autoincrement=True)
    min_text = Column('min_text', String, nullable=True)
    sent_count = Column('sent_count', Integer, nullable=True)
    cumsum_count = Column('cumsum_count', Integer, nullable=True)
    
    def __repr__(self):
        return 'pk:{}, min_text:{}'.format(self.pk, self.min_text)
```

```python
def read_input(zip_data_file_name, data_file_name):
    with zipfile.ZipFile(zip_data_file_name).open(data_file_name) as f:
        for i, line in enumerate(f):
            yield line.strip()
            
zip_data_file_name = 'total_text.zip'
data_file_name = 'total_text.txt'
documents = list(read_input(zip_data_file_name, data_file_name))
```
